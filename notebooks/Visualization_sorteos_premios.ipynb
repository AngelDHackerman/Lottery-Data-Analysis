```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load Datasets 
sorteos_df = pd.read_csv('../Data/downloaded/sorteos_export.csv')
premios_df = pd.read_csv('../Data/downloaded/premios_export.csv')
```

```python
# Join datasets by 'numero_sorteo'
data_combined_df = pd.merge(premios_df, sorteos_df, on='numero_sorteo', how='inner')

# Removing column "id"
data_combined_df.drop(columns=['id'], inplace=True)

# Save the new dataset for latter analysis in AWS QuickSight
data_combined_df.to_csv('../Data/processed/data_combined.csv', index=False)
```

### Data imputation

```python
# replace the null values with "N/A"
"""
Vendedor, ciudad and departamento are categorical variables, 
and filling them with N/A does not affect the analysis.
"""

data_combined_df['vendedor'] = data_combined_df['vendedor'].fillna("N/A")
data_combined_df['ciudad'] = data_combined_df['ciudad'].fillna("N/A")
data_combined_df['departamento'] = data_combined_df['departamento'].fillna("N/A")

data_combined_df
```
