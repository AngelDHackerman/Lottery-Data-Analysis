### 1. First Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# set the style
plt.style.use("ggplot")
sns.set_theme()
```

```python
# Load dataset
file_path = "../Data/downloaded/sorteos_export.csv"
df = pd.read_csv(file_path)
```

### 2. Explore The Data

### 3. Distribution of draw dates 

```python
df['fecha_sorteo'] = pd.to_datetime(df['fecha_sorteo']) # convert to datetime type
df['fecha_caducidad'] = pd.to_datetime(df['fecha_caducidad'])
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['fecha_sorteo'], bins=35, kde=True)
plt.title("Distribution of Draw Dates")
plt.xlabel("Date of draw")
plt.ylabel("Amount of draws")
plt.show()
```

### 4. Prize Analysis

```python
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# First place
sns.histplot(df['primer_premio'], bins=20, ax=ax[0], kde=True)
ax[0].set_title("Distribution of First Place")

# Second place
sns.histplot(df['segundo_premio'], bins=20, ax=ax[1], kde=True)
ax[1].set_title("Distribution of Second Place")

# Third place
sns.histplot(df['tercer_premio'], bins=20, ax=ax[2], kde=True)
ax[2].set_title("Distribution of Third Place")

plt.show()
```
